2. Write a Python program to reverse a word after accepting the input from the user.

Input word: ineuron
Output: norueni

```python
#question1
for i in range(1,6):
    for j in range(i):
        print('*',end = '')
    print('\n')
for i in range(4,0,-1):
    for j in range(i):
        print('*',end = '')
    print('\n')
```

```python
class reversing:
    def __init__(self,name):
        self.name=name
    def rev(self):
        return self.name[::-1]
```

```python
reversing('ineuron').rev()
```
